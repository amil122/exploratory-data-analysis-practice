```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
 df_test = pd.read_csv('dataset/black_friday/test.csv')
```

```python
df_train =  pd.read_csv('dataset/black_friday/train.csv')
```

```python
df = pd.concat([df_train, df_test], ignore_index=True)
```

```python
df
```

```python
df.info()
```

```python
df.drop('User_ID',axis = 1, inplace = True)
```

```python
df
```

```python
df.isna().sum()
```

# Handling categorical value gender into 0 and 1

```python
df['Gender'] = df['Gender'].apply(lambda x: 0 if x == 'M' else 1 if x == 'F' else x)
```

```python
df
```

# Handling categorical value age

```python
df['Age'].unique()
```

```python
df['Age'] = df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df
```

```python
df['City_Category'].unique()
```

```python
df_city = pd.get_dummies(df['City_Category'], drop_first = True)
```

```python
df_city
```

```python
df = pd.concat([df,df_city], axis = 1)
df
```

```python
df.drop('City_Category', axis = 1, inplace = True)
```
